###누락된 값 처리하기
1. 누락된 값 개수 확인하기: isna()메서드와 sum() 메서드
+ 누락되지 않은 값은 notna() 메서

```python
ns_book4.isna().sum()
```

```python
ns_book4.notna().sum()
```

2. 누락된 값으로 표시하기

```python
ns_book4.loc[0,'도서권수'] = None #loc[행,열]
ns_book4['도서권수'].isna().sum()
```

```python
ns_book4.head(2)
```

데이터 타입 바꾸기 : astype() 메서드
astype({열 이름:데이터 타입})

```python
ns_book4.loc[0,'도서권수']=1
ns_book4 = ns_book4.astype({'도서권수':'int32','대출건수':'int32'})
ns_book4.head(2)
```

문자열에 저장할 수 있는 열에 NaN 표시하기


```python
#None 입력하기
ns_book4.loc[0,'부가기호']=None
ns_book4.head(2)
```

```python
#None 대신 NaN으로 표시하기 (정수형 열에서)
import numpy as np
ns_book4.loc[0,'부가기호']=np.nan
ns_book4.head(2)
```

누락된 값 바꾸기(1): loc,fillna() 메서드
누락된 값을 NaN이 아니라 빈 문자열('')로 바꾸기

```python
set_isbn_na_rows=ns_book4['세트 ISBN'].isna() #누락된 값을 불리언 배열로 반환
ns_book4.loc[set_isbn_na_rows,'세트 ISBN']='' #누락된 값을 빈문자열로 바꿈
ns_book4['세트 ISBN'].isna().sum() #누락된 값이 몇 개인지 셈
```

더 편리한 방법: fillna() 메서드
모든 NaN을 '없음' 문자열로 바꾸기
-> isna() 메서드를 연결해 NaN의 개수를 셈

```python
ns_book4.fillna('없음').isna().sum()
```

```python
# 특정 열만 바꾸는 방법
ns_book4['부가기호'].fillna('없음').isna().sum()
```

```python
#특정 열을 바꾸고 전체 데이터 프레임을 반환하는 법
ns_book4.fillna({'부가기호':'없음'}).isna().sum()
```

누락된 값 바꾸기(2): replace() 메서드

```python
# 하나만 바꿀때 replace(원래 값, 새로운 값)
ns_book4.replace(np.nan,'없음').isna().sum()
```

```python
# 여러 개를 바꿀때 replace([원래값1,원래 값2],[새로운 값1, 새로운 값2]) 리스트 형식으로
ns_book4.replace([np.nan,'2021'],['없음','21']).head(2)
```

```python
#리스트 대신 딕셔너리 형식으로
ns_book4.replace({np.nan:'없음', '2021':'21'}).head(2)
```

열마다 다른 값으로 바꿀때
replace({열 이름 : 원래값}, 새로운 값)

```python
ns_book4.replace({'부가기호':np.nan},'없음').head(2)
```

```python
#중첩된 딕셔너리로 전달
# 열 이름과 변경 전후의 값 ({열 이름:{원래값1:새로운 값1}})
ns_book4.replace({'부가기호':{np.nan:'없음'},'발행년도':{'2021':'21'}}).head(2)
```

##정규 표현식
숫자 찾기 : \d
네자리 연도는 \d\d\d\d
그룹으로 묶을때 괄호를 사용
뒤에 두자리를 하나의 그룹으로 묶을 때 \d\d(\d\d)
정규표현깃을 사용한다는 의미: regex 매개변수 옵션을 True로 지정

```python
ns_book4.replace({'발행년도':{r'\d\d(\d\d)':r'\1'}},regex=True)[100:102]
#정규표현식 앞 r 문자는 정규표현식과 다른 문자열을 구분하기 위한 접두사
```

```python
#정규 표현식이 반복되면 중괄호로 개수 지정 가능
# \d{2}==\d\d
ns_book4.replace({'발행년도':{r'\d{2}(\d{2})':r'\1'}},regex=True)[100:102]
```
